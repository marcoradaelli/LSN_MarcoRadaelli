# Esecuzione di test sul generatore di numeri casuali #
In questa prima lezione abbiamo eseguito alcuni test per verificare la bontà statistica dei risultati ottenuti mediante il generatore di numeri casuali. In particolare, si verifica la proprietà di uniformità del generatore controllando il fatto che la sua media tenda al valore atteso di 0.5, e che la sua deviazione standard tenda al valore atteso di $\frac{1}{12}$. Viene poi eseguito un altro test con il metodo del $\chi^2$, per verificare che i valori siano effettivamente equidistribuiti all'interno dell'intervallo considerato.

## Test della media ##
Abbiamo inizialmente verificato che il valore medio del generatore di numeri casuali utilizzato fosse effettivamente 0.5, come atteso da un generatore che dia valori casuali uniformi nell'intervallo $[0,1)$. Per farlo, abbiamo generato $10^5$ numeri casuali. Poi abbiamo eseguito misure del valor medio a blocchi di 100 valori. Mostro i risultati di seguito.

```python
# Questo codice apre il file dei dati, li legge e li inserisce in un grafico apposito.
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig=plt.figure(figsize=(15, 13))                     #sx: larghezza; dx: Lunghezza

x, f, error = np.loadtxt("./Esercizio01.1/data.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.xlabel('Numero blocchi')
plt.ylabel('Media')
plt.grid()
plt.title("Media del RNG")

plt.show()
```

Si osserva che il valore si attesta sempre più vicino a 0.5 (che è il valor medio atteso) all'aumentare del numero dei lanci. L'errore tende a ridursi all'aumentare del numero di blocchi. Il valore di 0.5 è all'interno di una deviazione standard dal valore atteso, quindi riteniamo sia compatibile con le misure eseguite.

## Test della varianza ##
Ci si attende che la deviazione standard, calcolata con la formula
\begin{equation}
    \sigma^2 = \int_0^1{( x-\frac{1}{2} ) ^2}
\end{equation}
dia il valore analiticamente calcolabile di $\frac{1}{12}$. Valuto, con un algoritmo di fatto identico a quello utilizzato per il test della media, se questo è verificato per il nostro RNG.

```python
# Questo codice apre il file dei dati, li legge e li inserisce in un grafico apposito.
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig=plt.figure(figsize=(15, 13))                     #sx: larghezza; dx: Lunghezza

x, f, error = np.loadtxt("./Esercizio01.1b/sigma.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.xlabel('Numero blocchi')
plt.ylabel('Media')
plt.grid()
plt.title("Varianza del RNG")

plt.show()
```

Il grafico mostra che la deviazione standard tende effettivamente al valore atteso ($\frac{1}{12} = 0.8333$) con un errore che va a ridursi, dopo le oscillazioni iniziali.

## Test del $\chi^2$##
Usiamo questo test per verificare l'effettiva uniformità del generatore nell'intervallo considerato. In particolare, suddivido l'intervallo in $M=100$ sotto-intervalli tutti della stessa ampiezza. Poi comincio a lanciare $N=10^4$ numeri casuali con il generatore, e conto quanti di questi finiscono in ciascun sotto-intervallo. A questo punto, in linea teorica in ogni intervallo dovrei avere $\bar{x_i}=\frac{N}{M}=100$ valori. Per verificare quanto questa ipotesi sia realistica, uso il test del $\chi^2$, che viene calcolato come: $$\chi^2=\sum_{i=1}^{i=M} \frac{(x_i-\bar{x_i})^2}{\bar{x_i}}$$
L'esperimento viene ripetuto 100 volte, ottenendo quindi 100 valori diversi per il $\chi^2$, che riporto nel grafico.

```python
# Questo codice apre il file dei dati, li legge e li inserisce in un grafico apposito.
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig=plt.figure(figsize=(15, 13))                     #sx: larghezza; dx: Lunghezza

x, f = np.loadtxt("./Esercizio01.1c/chisq.dat", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,f)
plt.xlabel('Iterazione')
plt.ylabel('Chi quadrato')
plt.grid()
plt.title("Chi quadrato del RNG")
plt.autoscale(False, 'y')
# Imposto i limiti per gli assi.
AxisLimits = [0,100,0,140]
plt.axis(AxisLimits)
plt.autoscale(True, 'x')

plt.show()
```

Osservo che il grafico mostra, come atteso, che il $\chi^2$ si attesta intorno al valore di $\frac{N}{M}=100$, seppure con considerevoli oscillazioni. Osservo anche che l'importo relativo di tali oscillazioni viene a ridursi significativamente aumentando il numero degli intervalli $M$ (e di conseguenza anche il numero dei lanci $N$ in modo da avere un adeguato numero di lanci per ogni intervallo). 
Infine, non osservo una significativa variazione della situazione eseguendo un numero molto elevato di prove (cioè aumentando il "numero di numeri" considerati): la proprietà di uniformità non viene persa nel prosieguo della sequenza. Possiamo quindi ritenere che il generatore di numeri casuali da noi utilizzato sia efficace nel generare numeri casuali uniformemente nell'intervallo $[0,1)$.

# Verifica del teorema del limite centrale#
Il teorema del limite centrale (CLT) afferma che la somma di variabili casuali estratte da pdf con varianza finita si distribuisce, nel limite di una somma infinita, come una curva gaussiana. Per verificare la validità di tale enunciato, estraggo numeri casuali da una distribuzione uniforme nell'intervallo $[0,1)$ e da una esponenziale con parametro $\lambda = 1$. Eseguo somme a gruppi di 1, poi di 2, di 10 e di 100. Per la distribuzione uniforme, i risultati sono questi.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

fig=plt.figure(figsize=(15, 13))                     #sx: larghezza; dx: Lunghezza
fig.subplots_adjust(hspace=0.4, wspace=0.5)

def f(x,a,devstd,media):  # defining the fitting function
    return a/np.sqrt(2*np.pi*devstd**2) * np.exp(-0.5*((x-media)/devstd)**2)

plt.subplot(2,2,1)
x, rho = np.loadtxt("./Esercizio01.2a/uniform_data_with_1.dat", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,rho)
plt.xlabel("x")
plt.ylabel("rho(x)")
plt.title('Uniforme con N=1')
AxisLimits = [0,100,0,140]
plt.axis(AxisLimits)
plt.autoscale(True, 'x')

plt.subplot(2,2,2)
plt.autoscale(True,'y')
x, rho = np.loadtxt("./Esercizio01.2a/uniform_data_with_2.dat", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,rho)
plt.xlabel("x")
plt.ylabel("rho(x)")
plt.title('Uniforme con N=2')

plt.subplot(2,2,3)
x, rho = np.loadtxt("./Esercizio01.2a/uniform_data_with_10.dat", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,rho)
plt.xlabel("x")
plt.ylabel("rho(x)")
plt.title('Uniforme con N=10')

plt.subplot(2,2,4)
x, rho = np.loadtxt("./Esercizio01.2a/uniform_data_with_100.dat", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,rho)
plt.xlabel("x") 
plt.ylabel("rho(x)")
plt.title('Uniforme con N=100')

p_opt, p_cov = curve_fit(f, x, rho, bounds=([1.2e4,1,40],[1.6e4,30,60]))
y_fit = f(x,p_opt[0],p_opt[1],p_opt[2])
plt.plot(x,y_fit) # plotting fitted function
print("optimized parameters [a,devstd,media] =")
print(p_opt)
print("parameters uncertainty =")
print(np.sqrt(np.diagonal(p_cov)))

plt.show()
```

Si osserva come al crescere di $N$ l'andamento della distribuzione delle somme sia sempre più simile ad una gaussiana. Il teorema del limite centrale appare quindi verificato. Questo è anche evidente dal fit gaussiano eseguito. Ci aspettiamo che la distribuzione delle somme tenda ad una gaussiana avente valor medio uguale al valore medio moltiplicato per il numero di valori sommati (quindi $0.5 \cdot 100 = 50$) e deviazione standard uguale alla deviazione standard moltiplicata per il numero di valori sommati sotto radice. 
Lo stesso procedimento si può applicare per una distribuzione esponenziale (in cui abbiamo posto il parametro $\lambda=1$). Si ottiene:

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

fig=plt.figure(figsize=(15, 13))                     #sx: larghezza; dx: Lunghezza
fig.subplots_adjust(hspace=0.4, wspace=0.5)

def f(x,a,devstd,media):  # defining the fitting function
    return a/np.sqrt(2*np.pi*devstd**2) * np.exp(-0.5*((x-media)/devstd)**2)

plt.subplot(2,2,1)
x, rho = np.loadtxt("./Esercizio01.2b/exponential_data_with_1.dat", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,rho)
plt.xlabel("x")
plt.ylabel("rho(x)")
plt.title('Esponenziale con N=1')

plt.subplot(2,2,2)
x, rho = np.loadtxt("./Esercizio01.2b/exponential_data_with_2.dat", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,rho)
plt.xlabel("x")
plt.ylabel("rho(x)")
plt.title('Esponenziale con N=2')

plt.subplot(2,2,3)
x, rho = np.loadtxt("./Esercizio01.2b/exponential_data_with_10.dat", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,rho)
plt.xlabel("x")
plt.ylabel("rho(x)")
plt.title('Esponenziale con N=10')

plt.subplot(2,2,4)
x, rho = np.loadtxt("./Esercizio01.2b/exponential_data_with_100.dat", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,rho)
plt.xlabel("x")
plt.ylabel("rho(x)")
plt.title('Esponenziale con N=100')

p_opt, p_cov = curve_fit(f, x, rho, bounds=([1.2e4,1,80],[1.6e4,30,120]))
y_fit = f(x,p_opt[0],p_opt[1],p_opt[2])
plt.plot(x,y_fit) # plotting fitted function
print("optimized parameters [a,devstd,media] =")
print(p_opt)
print("parameters uncertainty =")
print(np.sqrt(np.diagonal(p_cov)))

plt.show()
```

# Somma di una distribuzione a varianza infinita #
Il CLT ha tra le sue ipotesi che la varianza della distribuzione sia finita. Questo non è valido per alcune applicazioni di interesse fisico, come nel caso della distribuzione di Cauchy-Lorentz (nel nostro caso consideriamo i parametri $\mu=0$, $\Gamma=1$). In questo caso, il limite delle somme non è infatti una gaussiana, ma un'altra distribuzione. Nel caso della distribuzione di Cauchy-Lorentz si può dimostrare che la distribuzione limite è ancora una distribuzione di Cauchy-Lorentz. Lo verifico numericamente:

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig=plt.figure(figsize=(15, 13))                     #sx: larghezza; dx: Lunghezza
fig.subplots_adjust(hspace=0.4, wspace=0.5)

plt.subplot(2,2,1)
x, rho = np.loadtxt("./Esercizio01.2c/cauchy_data_with_1.dat", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,rho)
plt.xlabel("x")
plt.ylabel("rho(x)")
plt.title('Cauchy-Lorentz con N=1')

plt.subplot(2,2,2)
x, rho = np.loadtxt("./Esercizio01.2c/cauchy_data_with_2.dat", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,rho)
plt.xlabel("x")
plt.ylabel("rho(x)")
plt.title('Cauchy-Lorentz con N=2')

plt.subplot(2,2,3)
x, rho = np.loadtxt("./Esercizio01.2c/cauchy_data_with_10.dat", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,rho)
plt.xlabel("x")
plt.ylabel("rho(x)")
plt.title('Cauchy-Lorentz con N=10')

plt.subplot(2,2,4)
x, rho = np.loadtxt("./Esercizio01.2c/cauchy_data_with_100.dat", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,rho)
plt.xlabel("x")
plt.ylabel("rho(x)")
plt.title('Cauchy-Lorentz con N=100')
plt.show()
```

# Esperimento dell'ago di Buffon #
L'esperimento dell'ago di Buffon, da cui si può in particolare ricavare una stima per via statistica del valore di $\pi$, è un interessante esempio di simulazione. Immagino di lanciare moltissime volte un ago di lunghezza nota $L$ su un foglio dotato di righe a distanza $d$ l'una dall'altra, con $L < d$ in modo da evitare aghi che intersechino più di una riga. A questo punto si può dimostrare che, detto $n_{throw}$ il numero di lanci e $n_{hit}$ il numero di volte in cui l'ago interseca una riga, si ha:
$$
\pi_{estimated} = \lim_{n_{throw} \to \infty} 2 \cdot \frac{L n_{throw}}{d n_{hit}}
$$
Ho tentato di realizzare questa simulazione numericamente. Visto che ovviamente non posso usare il valore di $\pi$ (che è proprio quello che vorrei calcolare) invece di estrarre l'angolo formato dall'ago uniformemente nell'intervallo $[0, \pi)$ estraggo un punto come coppia di valori $(x,y)$ all'interno della circonferenza unitaria, poi calcolo l'angolo tramite $\alpha = \arctan \frac{y}{x}$.

Riporto nel seguente grafico i dati ottenuti, calcolando l'errore con il consueto metodo dei blocchi.

```python
# Questo codice apre il file dei dati, li legge e li inserisce in un grafico apposito.
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig=plt.figure(figsize=(15, 13))                     #sx: larghezza; dx: Lunghezza

x, f, error = np.loadtxt("./Esercizio01.3/data.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.xlabel('Numero blocchi')
plt.ylabel('Stima di pi greco')
plt.grid()
plt.title("Esperimento dell'ago di Buffon")

plt.show()
```

Il valore ottenuto non è compatibile con $\pi$. In particolare, noto che si ha una fortissima dipendenza del risultato da parecchi parametri, in particolare dalle dimensioni del foglio su cui si lancia. In generale, provando a variare i parametri, sembra che l'algoritmo abbia la tendenza a sovrastimare il valore di $\pi$.
