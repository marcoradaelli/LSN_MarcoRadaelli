# Simulazione di Meccanica Statistica Quantistica #
Scopo del presente laboratorio è individuare, usando metodi di Quantum Montecarlo, una espressione approssimata del ground state quantistico per un potenziale a doppia buca della forma:
$$
V(x) = x^4 - \frac{5}{2}x^2
$$
Per questo problema non è disponibile una soluzione analitica, quindi è necessario affrontarlo con tecniche di calcolo numerico.

## Variational Montecarlo ##
La prima tecnica con cui si è attaccato il problema è quella del <i> variational Montecarlo </i>. Si propone una funzione d'onda prototipo $\Psi_T$ di cui si è certi che non sia ortogonale alla wave function del ground state sulla base di argomenti di simmetria, nel nostro caso una somma di gaussiane:
$$
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$
A questo punto si vogliono individuare i valori ottimali dei parametri $\mu$ e $\sigma$. D'altra parte, siamo certi che l'energia del ground state sia inferiore a quella di qualunque stato, quindi il problema si riduce a trovare $\mu$ e $\sigma$ che minimizzino l'energia, calcolata tramite l'integrale:
$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} 
$$
Tale integrale viene calcolato con una tecnica Montecarlo tramite l'algoritmo di Metropolis, osservando che si può equivalentemente calcolare come:
$$
\langle {\hat H} \rangle_T =  
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}
$$
Quindi si usa l'algoritmo di Metropolis per campionare il modulo quadro della funzione d'onda $\Psi_T$ (che è certamente una "buona" distribuzione di probabilità) e poi si usano i punti ottenuti per calcolare l'integrale per valori dati di $\mu$ e $\sigma$.

Resta a questo punto il problema di trovare i valori di tali parametri che minimizzano l'energia. La tecnica che ho usato è quella di una "rete" di valori di $\mu$ e $\sigma$. I parametri della rete sono specificabili negli appositi campi dell'inputfile. Viene calcolato il valore di aspettazione dell'energia a tutti i punti della rete. Se si attiva l'opzione di ricerca automatica, il programma a questo punto individua automaticamente il minimo di energia su questa rete e "ricentra" la ricerca, esplorando meglio quell'area, il tutto per un numero di volte specificato nell'inputfile. 
Eseguendo numerose volte la procedura di ricentramento, le migliori stime dei parametri sono state: $\mu = (0.805389 \pm 0.000005) $ ; $\sigma = (0.610831 \pm 0.000005) $. Con tali valori si ottiene una stima di energia di $\langle {\hat H} \rangle_T=-0.446261$.

Il seguente grafico mostra la convergenza dell'integrale di energia in funzione del numero di blocchi considerati per i parametri $\mu$ e $\sigma$ ottimali, i cui valori sono stati dati in precedenza.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig=plt.figure(figsize=(15, 8))                     #sx: larghezza; dx: Lunghezza

N, E, err = np.loadtxt("./Esercizio8.01/data.dat", usecols=(0,1,2), delimiter='   ', unpack='true')

plt.errorbar(N,E,yerr=err)
plt.grid(True)
plt.xlabel("Numero di blocchi")
plt.ylabel("Valore aspettazione dell'energia")
plt.title("Valore di aspettazione energia per parametri ottimali")

plt.show()
```

## Path Integral Ground State (PIGS)  ##
Infine, uso il codice fornito per calcolare il ground state usando il più recente ed avanzato algoritmo PIGS. Dapprima, uso come funzione di partenza la funzione $\Psi_T(x) = 1$. In questo caso ottengo un modulo quadro della wave function:

Riporto ora i grafici ottenuti in precedenza sugli stessi assi, in modo da poter eseguire un confronto.

```python
x1, N1 = np.loadtxt("./Esercizio8.02/istogramma.dat", usecols=(0,1), delimiter='   ', unpack='true')
x2, N2 = np.loadtxt("./QMC_1D/probability_from_identity.dat", usecols=(0,1), delimiter=' ', unpack='true')
x3, N3 = np.loadtxt("./QMC_1D/probability_from_optimized.dat", usecols=(0,1), delimiter=' ', unpack='true')

fig=plt.figure(figsize=(15, 8))                     #sx: larghezza; dx: Lunghezza

plt.plot(x1,N1, color = 'red')
plt.plot(x2,N2, color = 'green')
plt.plot(x3,N3, color = 'blue')
plt.grid(True)

plt.title("Modulo quadro della funzione d'onda")
plt.xlabel("x")
plt.ylabel("Modulo quadro w.f.")

plt.show()
```

Si osserva che l'accordo tra le predizioni ottenute con i due PIGS è pressoché ovunque perfetto. Invece l'accordo con il modulo quadro della w.f. ottenuto tramite il metodo Variational Montecarlo è buono lontano dai punti stazionari del modulo quadro stesso, dove si osservano discrepanze evidenti. Questo può essere dovuto al fatto che nell'intorno dei punti stazionari la "specificità" della w.f. è più marcata, e quindi una sua approssimazione con una funzione di prova, seppur ottimizzata per minimizzare l'energia, è peggiore. Infine, posso paragonare i risultati ottenuti con quelli che si hanno da un altro metodo numerico basato sulla discretizzazione dell'equazione di Schroedinger e sulla sua trasformazione in una ordinaria equazione agli autovalori per vettori:

```python
import numpy as np
import matplotlib.pyplot as plt

def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

print("Ground state energy: ", E[0])
print("1st excited state energy: ", E[1])
print("2nd excited state energy: ", E[2])

# Plot a few things
fig=plt.figure(figsize=(15, 8))                     #sx: larghezza; dx: Lunghezza
plt.plot(x,(psi[0])**2, color = 'black')

x1, N1 = np.loadtxt("./Esercizio8.02/istogramma.dat", usecols=(0,1), delimiter='   ', unpack='true')
x2, N2 = np.loadtxt("./QMC_1D/probability_from_identity.dat", usecols=(0,1), delimiter=' ', unpack='true')
x3, N3 = np.loadtxt("./QMC_1D/probability_from_optimized.dat", usecols=(0,1), delimiter=' ', unpack='true')

plt.plot(x1,N1, color = 'red')
plt.plot(x2,N2, color = 'green')
plt.plot(x3,N3, color = 'blue')
plt.grid(True)

plt.title("Modulo quadro della funzione d'onda")
plt.xlabel("x")
plt.ylabel("Modulo quadro w.f.")

plt.show()
```

Anche in questo caso si osserva che le predizioni PIGS sono praticamente sovrapposte a quelle del metodo di discretizzazione, mentre le predizioni del Variational Montecarlo si discostano nei pressi dei punti stazionari.

Conviene usare come funzione di input per l'algoritmo PIGS quella ottimizzata ottenuta dal Variational Montecarlo rispetto ad una funzione a caso (seppur non ortogonale al ground state reale) come la $\Psi_T(x)=1$ ? Considero i grafici dell'energia potenziale lungo il polimero PIGS, avviati con lo stesso file di input ma le due diverse funzioni di partenza:

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig=plt.figure(figsize=(15, 8))                     #sx: larghezza; dx: Lunghezza

N, E, err = np.loadtxt("./QMC_1D/potential_from_identity.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
N2, E2, err2 = np.loadtxt("./QMC_1D/potential_from_optimized.dat", usecols=(0,1,2), delimiter=' ', unpack='true')

plt.errorbar(N,E,yerr=err, color='red')
plt.errorbar(N2,E2,yerr=err2, color='blue')
plt.grid(True)
plt.xlabel("Posizione polimero PIGS")
plt.ylabel("Energia potenziale")
plt.title("Energia potenziale lungo il polimero")

plt.show()
```

Si osserva come, partendo da una w.f. di prova più "simile" al ground state, l'energia potenziale si attesti fin dagli estremi del polimero (dove il ket e il bra sono più simili a quelli di prova) su valori vicini a quelli che raggiunge al centro del polimero. Invece usando una w.f. di prova molto diversa (come la $\Psi_T(x)=1$) l'energia inizialmente deve scendere di parecchio prima di attestarsi sul valore corretto.

# Path Integral Monte Carlo (PIMC) e dipendenza dalla temperatura #

Come cambia la situazione al variare della temperatura? Il programma di Quantum Montecarlo che stiamo utilizzando permette di determinarlo con un algoritmo detto Path Integral Monte Carlo (PIMC). Riporto di seguito i grafici ottenuti per la probabilità di trovare la particella in un certo punto. Osservo che non è più possibile parlare di wave function: quando inseriamo una trattazione termodinamica non abbiamo più stati puri ma stati misti (miscele classiche di stati quantistici), e quindi uno stato non è più ben rappresentato da un ket (o da un suo sviluppo sulla base delle coordinate, cioè dalla funzione d'onda).

```python
x1, N1 = np.loadtxt("./QMC_1D/probability_PIMC_1.00.dat", usecols=(0,1), delimiter=' ', unpack='true')
x2, N2 = np.loadtxt("./QMC_1D/probability_PIMC_1.25.dat", usecols=(0,1), delimiter=' ', unpack='true')
x3, N3 = np.loadtxt("./QMC_1D/probability_PIMC_0.75.dat", usecols=(0,1), delimiter=' ', unpack='true')
x4, N4 = np.loadtxt("./QMC_1D/probability_from_optimized.dat", usecols=(0,1), delimiter=' ', unpack='true')

fig=plt.figure(figsize=(15, 8))                     #sx: larghezza; dx: Lunghezza

plt.plot(x1,N1, color = 'red')
plt.plot(x2,N2, color = 'green')
plt.plot(x3,N3, color = 'blue')
plt.plot(x4,N4, color = 'black')
plt.grid(True)

plt.title("Probabilità")
plt.xlabel("x")
plt.ylabel("Probabilità")

plt.show()
```
