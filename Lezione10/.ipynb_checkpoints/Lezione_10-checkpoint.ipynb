# Algoritmo di simulated annealing per il problema TSP #
Obiettivo della presente lezione è l'uso di un algoritmo di simulated annealing per la risoluzione numerica del problema del commesso viaggiatore (TSP). Tale algoritmo si basa sull'idea di trattare il problema come un sistema termodinamico di cui la soluzione sia il ground state. Quella che per l'algoritmo genetico era la funzione costo (cioè, nel caso specifico, la lunghezza del percorso prescelto) diventa nel caso del simulated annealing una sorta di hamiltoniana fittizia. A questo punto si tratta il sistema come se fosse un ensemble canonico (quindi a temperatura fissata), dunque la probabilità di una certa configurazione che comporti una "energia" di un certo valore è data dal peso di Boltzmann. La distribuzione può quindi facilmente essere campionata usando l'algoritmo Metropolis a temperatura fissata. 

L'idea del simulated annealing è quella di raggiungere il ground state tramite un progressivo abbassamento della temperatura: si parte da una temperatura alta, si lascia che l'algoritmo Metropolis campioni un po' il peso di Boltzmann a quella temperatura, poi si scende a una temperatura più bassa e così via. 

Le mosse Metropolis considerate sono quelle che erano state scritte come mutazioni per l'algoritmo genetico:
<ul>
    <li>scambio di due città scelte a caso</li>
    <li>spostamento tutte assieme di $n$ città contigue di $m$ posizioni</li>
    <li>permutazione di due blocchi di città, spezzando il percorso in una posizione $m$ scelta a caso</li>
</ul>

Si osserva che allo scendere della temperatura la probabilità di accettazione Metropolis delle mosse diminuisce significativamente: pian piano si è raggiunta una configurazione sempre più vicina al ground state, quindi qualunque mossa costa troppa energia. Questo si verifica per mosse dei primi due tipi. Resta invece alta la probabilità di accettazione della permutazione a blocchi. Ritengo che questo sia dovuto essenzialmente alla presenza, nel pattern di città considerato, di città tra loro molto vicine. Il costo energetico di una inversione di percorso fra queste città è ridottissimo.

Di seguito riporto i risultati ottenuti usando il simulated annealing.

```python
import numpy as np
import matplotlib.pyplot as plt

fig=plt.figure(figsize=(15, 8))  # Imposto le misure del grafico.

x1,y1 = np.loadtxt("./Es10.1/migliori.dat", usecols=(0,1), delimiter='   ', unpack='false')

plt.plot(x1,y1)
plt.grid(True)

plt.title("Convergenza con simulated annealing su circonferenza")
plt.xlabel("Beta")
plt.ylabel("Lunghezza stimata")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig=plt.figure(figsize=(8, 8))  # Imposto le misure del grafico.

x1,y1 = np.loadtxt("./Es10.1/movimenti.dat", usecols=(0,1), delimiter='   ', unpack='false')

plt.plot(x1,y1)
plt.grid(True)

plt.title("Simulated annealing: percorso finale")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```
