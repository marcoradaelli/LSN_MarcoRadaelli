# Simulazione di una opzione europea #
Utilizzo un metodo Montecarlo per il calcolo del prezzo di una opzione europea. Questo tipo di opzione e' particolarmente interessante, perche' per il prezzo esiste un calcolo analitico, basato sulla formula di Black e Scholes, che quindi ci da' un termine di paragone per il risultato ottenuto via simulazione numerica.

### Simulazioni one-step ###
Visto che il campionamento di un moto browniano effettuato e' esatto e non approssimato, non e' necessario procedere a piccoli passi: contrariamente a quanto accade p.es. nella soluzione di equazioni differenziali infatti non vi e' un errore influenzato dalla dimensione del passo di integrazione. Una prima tecnica utilizzata e' quindi quella di simulare un gran numero di volte la variazione dei prezzi con una sola valutazione del simulatore. Ovviamente questa tecnica e' piu' semplice dal punto di vista della programmazione e anche molto piu' veloce all'esecuzione.
I parametri considerati sono stati:
<ul>
    <li>asset price a $t=0$: $S(0)=100$</li>
    <li>delivery time: $T=1$</li>
    <li>strike price: $K=1004$</li>
    <li>risk-free interest rate: $r=0.1$</li>
    <li>volatility: $\sigma=0.25$</li>
</ul>
Provo poi a valutare come varia il prezzo al variare di questi parametri. Riporto nei grafici sotto cosa succede con i parametri:
<ul>
    <li>(1) Nelle condizioni iniziali</li>
    <li>(2) Portando $T$ a 3</li>
    <li>(3) Portando $K$ a 110</li>
    <li>(4) Portando $\sigma$ a 0.3</li>
</ul>

```python
# Questo codice apre il file dei dati, li legge e li inserisce in un grafico apposito.
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig=plt.figure(figsize=(15, 13))  #sx: larghezza; dx: Lunghezza
fig.subplots_adjust(hspace=0.2, wspace=0.2)

plt.subplot(2,2,1)
x, f, error = np.loadtxt("./Esercizio03_call_onestep/profits_T1.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error,color='r')
plt.xlabel('Numero blocchi')
plt.ylabel('Prezzo opzione')
plt.grid()
plt.title("Prezzo di una opzione call europea - 1")

plt.subplot(2,2,2)
x, f, error = np.loadtxt("./Esercizio03_call_onestep/profits_T3.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error,color='r')
plt.xlabel('Numero blocchi')
plt.ylabel('Prezzo opzione')
plt.grid()
plt.title("Prezzo di una opzione call europea - 2")

plt.subplot(2,2,3)
x, f, error = np.loadtxt("./Esercizio03_call_onestep/profits_K110.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error,color='r')
plt.xlabel('Numero blocchi')
plt.ylabel('Prezzo opzione')
plt.grid()
plt.title("Prezzo di una opzione call europea - 3")

plt.subplot(2,2,4)
x, f, error = np.loadtxt("./Esercizio03_call_onestep/profits_sigma03.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error,color='r')
plt.xlabel('Numero blocchi')
plt.ylabel('Prezzo opzione')
plt.grid()
plt.title("Prezzo di una opzione call europea - 4")

plt.show()
```

Il valore ottenuto e' compatibile con quello atteso secondo le formule di Black-Scholes.

### Comportamento al variare di $\sigma$ ###
Vorrei ora capire cosa succede se cambio il valore di $\sigma$. Per farlo, plotto il valore del prezzo finale in funzione di $\sigma$ su una call europea. Sovrappongo il risultato con quello teorico dato dalle formule di Black-Scholes:

```python
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy.special import erf
from scipy import log, exp, sqrt

fig=plt.figure(figsize=(15, 8))  #sx: larghezza; dx: Lunghezza

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    #P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C

call = []
sigmaArr=[]
sigma0=0.01;
sigma_step=0.01
# Inserisco i dati ottenuti in un array.
for i in range(5000):
    sigma = sigma0 + i/100.*sigma_step
    sigmaArr.append(sigma)
    call.append(black_scholes(S0, K, T, r, sigma))

# A questo punto plotta il tutto.
    
    
x, f = np.loadtxt("./Esercizio03_call_sigma/prices.dat", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,f,color='r')
plt.plot(sigmaArr,call)
plt.xlabel("Volatilita'")
plt.ylabel('Costo opzione asset')
plt.grid()
plt.title("Costo opzione al variare della volatilita'")

    
plt.show()
```

Osservo che il valore trovato per il prezzo dell'opzione sale in maniera circa rettilinea al salire del valore della volatilita', salvo per le volatilita' piu' basse. Posso ritenere che all'aumentare della volatilita' aumenti il possibile guadagno ottenuto, e quindi il costo dell'opzione.
